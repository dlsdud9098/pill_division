```python

pill_image_list = []
pill_images = glob('./pill_images/*')
i = 1
for pill in pill_images:
    idx = os.path.basename(pill)[:-4]
    img = Image.open(pill)
    # numpy 배열로 변경하기
    x = np.array(img)

    pill_image_dict = {
        '품목일련번호': idx,
        '이미지': x
    }
    pill_image_list.append(pill_image_dict)
pill_image_csv = pd.DataFrame(pill_image_list)
pill_image_csv
```
